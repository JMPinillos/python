Modifique los valores del atributo model, de tal manera que los valores `x Series` pasen a ser `Serie x`, siendo *x* un número entre 1 y 9.

```python
df_mod = df
diccionario = dict()

for i in range(1,10):
    diccionario[" " + str(i) + " Series"] = "Series " + str(i)

df_mod['model'] = df['model'].replace(diccionario)

df_mod
```

Calcule para cada registro el número medio de millas recorridas cada año.

```python
from datetime import datetime

# Obtención del año actual
actualYear =  datetime.now().year

for indice, registro in df.iterrows():
    antique = actualYear - int(registro["year"])
    halfMiles = registro["mileage"] / antique
    print(halfMiles)
```
